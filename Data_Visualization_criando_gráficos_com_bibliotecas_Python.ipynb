#1 CONHECENDO A BIBLIOTECA MATPLOTLIB

## 1.1 Conhecendo os dados

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv'
```

```python
df = pd.read_csv(url)
```

```python
df
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
brasil_dict = {
    'ano': brasil.index.tolist(),
    'imigrantes': brasil.values.tolist()
}
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

## 2.6 visualizando dados de vendas de diferentes lojas

```python
lojas = ['A', 'B', 'C', 'D']
```

```python
vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df_vendas = pd.DataFrame(vendas_2022, index=lojas)
```

```python
fig_vendas, axs_vendas = plt.subplots(2, 2, figsize=(14, 8))
plt.subplots_adjust(wspace=0.3, hspace=0.4)
fig_vendas.suptitle('Vendas no período de janeiro a dezembro de 2022 nas lojas A,B,C e D')

axs_vendas[0, 0].plot(df_vendas.loc['A'])
axs_vendas[0, 0].set_title('Vendas na loja A')
axs_vendas[0, 1].plot(df_vendas.loc['B'])
axs_vendas[0, 1].set_title('Vendas na loja B')
axs_vendas[1, 0].plot(df_vendas.loc['C'])
axs_vendas[1, 0].set_title('Vendas na loja C')
axs_vendas[1, 1].plot(df_vendas.loc['D'])
axs_vendas[1, 1].set_title('Vendas na loja D')

for ax in axs_vendas.flat:
    ax.set_xlabel('Mês')
    ax.set_ylabel('Número de vendas')
    ax.grid()

ymin = 0
ymax = 500

for ax in axs_vendas.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```
